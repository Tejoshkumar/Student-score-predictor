```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Load the dataset from working URL
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
data = pd.read_csv(url)

# Step 3: View the data
print("ðŸ“Š First 5 rows of the dataset:")
print(data.head())

# Step 4: Plot the data
plt.scatter(data['Hours'], data['Scores'])
plt.xlabel('Hours Studied')
plt.ylabel('Score (%)')
plt.title('Hours vs Scores')
plt.show()
# Step 5: Split data into input (X) and output (y)
X = data[['Hours']]  # input features (2D)
y = data['Scores']   # output labels

# Step 6: Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Step 7: Predict score for a new input
hours = 9.25
predicted_score = model.predict([[hours]])
print(f"\nðŸ“ˆ If a student studies for {hours} hours, the predicted score is {predicted_score[0]:.2f}")
```
